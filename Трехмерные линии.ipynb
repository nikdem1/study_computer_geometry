```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np           
```

Будем рисовать различные кривые, в том числе и пространственные. Уравнения кривых будут в параметрическом виде.

###Винтовая линия
\begin{equation}
    \left\{
    \begin{aligned}
        &x(t) = 3a\cos(t)\\
        &y(t) = 3a\sin(t)\\
        &z(t) = 4at
    \end{aligned}
    \right.
\end{equation}

```python
# Параметр
a = 0.3
t = [t for t in np.arange(0.0,20.0,0.1)]
x = 3.0*a*np.cos(t)
y = 3.0*a*np.sin(t)
z = [4*a*tt for tt in t]

fig = plt.figure(0)
ax = fig.gca(projection='3d')
ax.plot(x,y,z)
ax.set_title("Винтовая линия")
```

###Циклоида
\begin{equation}
    \left\{
    \begin{aligned}
        &x(t) = a(t - \sin{t})\\
        &y(t) = a(1 - \cos{t})
    \end{aligned}
    \right.
\end{equation}

```python
a = 0.4
x = a*(t - np.sin(t))
y = a*(1 - np.cos(t))

fig2 = plt.figure(1)
ax2 = fig2.add_subplot(1,1,1)
ax2.plot(x,y)
ax2.set_title("Циклоида")
```
