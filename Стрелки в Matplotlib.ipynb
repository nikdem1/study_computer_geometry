```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.style.use("../default_colors.mplstyle")
```

```python
# Точки
P = np.array([
    (1, 2), (2, 3), (4, 2), (1, 5)
])
# Координаты вектора
V = P[1:] - P[:-1]
V
```

```python
# Хотим соединить две точки стрелкой
P1 = np.array([1.5, 3])
P2 = np.array([3, 4.5])
# Надо найти координаты вектора P2 - P1
P12 = P2-P1
```

```python
fig01 = plt.figure(num=1, dpi=300)

ax01 = fig01.add_subplot(1, 1, 1)

# стиль стрелки можно объявить как словарь и передать в arrow 
# это позволит собрать все настройки в одном месте
style = dict(length_includes_head=True, width=0.005, head_width=0.05, color='k')

# Точки P1 и P2
# arrow принимает координаты x, y начала стрелки и dx, dy --- координаты вектора
# получается, что стрелка рисуется от (x, y) к (x+dx, y+dy)
ax01.arrow(*P1, *P12, **style)
# Рисуем точки
ax01.plot(*P1, ls='None', marker='o', markersize=4, color='red')
ax01.plot(*P2, ls='None', marker='o', markersize=4, color='red')

# Рисуем стрелки
ax01.arrow(*P[0], *V[0], **style)
ax01.arrow(*P[1], *V[1], **style)
ax01.arrow(*P[2], *V[2], **style)
# Рисуем точки
ax01.plot(*P.T, ls='None', marker='o', markersize=4)
```
